```python
import csv

# CSV dosyasının yolunu belirtin
dosya_yolu = 'handbags.csv'

# Farklı marka isimlerini saklamak için bir küme (set) oluşturun
marka_isimleri = set()

# CSV dosyasını açın ve okuyun
with open(dosya_yolu, 'r', encoding='utf-8') as dosya:
    csv_okuyucu = csv.DictReader(dosya)
    
    for satir in csv_okuyucu:
        marka_isimleri.add(satir['UPC'])

# Farklı marka isimlerinin sayısını alın
marka_sayisi = len(marka_isimleri)

print(f"Toplam {marka_sayisi} farklı marka ismi bulunuyor.")
```

```python
import csv

# CSV dosyasının yolunu belirtin
dosya_yolu = 'handbags.csv'

# "Brand Name" sütunundaki eşleşmelerin sayısını saklamak için bir sayaç başlatın
brand_name_sayac = 0

# CSV dosyasını açın ve okuyun
with open(dosya_yolu, 'r', encoding='utf-8') as dosya:
    csv_okuyucu = csv.DictReader(dosya)
    
    for satir in csv_okuyucu:
        if 'Brand Name' in satir and satir['Brand Name']:
            brand_name_sayac += 1

print(f"'Brand Name' sütununda toplam {brand_name_sayac} satır bulunuyor.")
```

```python
import pandas as pd

# CSV dosyasını oku
df = pd.read_csv("handbags.csv")

# Index sütunu ekleyin (varsayılan olarak 0'dan başlar)
df['Index'] = range(1, len(df) + 1)

# Kolon isimlerini düzenle
df = df.rename(columns={'CurrentColumnName1': 'YeniAd1', 'CurrentColumnName2': 'YeniAd2', 'CurrentColumnName3': 'YeniAd3'})

# Düzenlenmiş veriyi yeni bir CSV dosyasına kaydet
df.to_csv("düzenlenmiş_handbags.csv", index=False)

print("CSV dosyası düzenlendi ve yeni dosya oluşturuldu.")
```
